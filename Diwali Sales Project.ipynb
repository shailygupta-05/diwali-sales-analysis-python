## Exploratory Data Analysis (EDA)

```python
df.columns
```

### Gender Analysis

```python
# gender_wise total orders

sns.set(rc={'figure.figsize':(5,5)})

ax=sns.countplot(data=df,x='Gender',color='orange')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# gender_wise total sale-

gender_sales = df.groupby('Gender')['Amount'].sum()

plt.figure(figsize=(5,5))

plt.pie(gender_sales, labels=['Female', 'Male'],autopct='%1.1f%%',startangle=90)

plt.title('Sales by Gender')
plt.show()
```

#### Insight:
from the above graphs we can clearly see that most of the buyers are Female, contributing higher overall revenue.

```python
df.head()
```

## Age Analysis

```python
# agegroup_wise total orders

sns.set(rc={'figure.figsize':(10,5)})

ax=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# agegroup_wise total sale-

sales_age= df.groupby('Age Group')["Amount"].sum().sort_values(ascending=False).reset_index()

sns.barplot(x='Age Group',y='Amount',data=sales_age,color='Brown')
```

#### Insights:
from the above graphs we can see that most of the buyers are of the age group between 26-35 yrs female.

```python
df.head()
```

## State Analysis

```python
# top 5 state by orders-

sales_state=df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(5).reset_index()

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data=sales_state,x='State',y='Orders',color='green')
```

```python
# top 5 state by sales-

sales_state=df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(5).reset_index()

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
# bottom 5 state by sales-

sales_state=df.groupby('State')['Amount'].sum().sort_values(ascending=False).tail(5).reset_index()

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data=sales_state,x='State',y='Amount',color='red')
```

#### Insights:
from the above graphs we can see that most of the orders and sales are from Uttar Pradesh, Maharashtra and Karnataka , contributing overall revenue while the least contributing states are Jharkhand,Uttarakhand and Rajasthan.

```python
df.head()
```

## Product Category Analysis

```python
# product category by orders-

sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(data=df,x='Product_Category',color='brown')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# TOP 5 product category by orders-

category_sales=df.groupby('Product_Category')['Amount'].count().sort_values(ascending=False).head(5).reset_index()
sns.barplot(data=category_sales,x='Product_Category',y='Amount',color='orange')
```

```python
# TOP 5 product category by sales-

category_sales=df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(5).reset_index()
sns.barplot(data=category_sales,x='Product_Category',y='Amount',color='green')
```

```python
# bottom 5 product category by sales-

category_sales=df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).tail(5).reset_index()
sns.barplot(data=category_sales,x='Product_Category',y='Amount',color='red')
```

Insight: from the above graphs we can see that people spend more on food,clothing and buying electronics items while least buying products are decor,pet care,and hand and power tools.

```python
df.head()
```

#### Occupation Analysis -

```python
# occupation wise total orders count-

sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(data=df,x='Occupation',color='blue')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# occupation wise total sales-

sales_occupation=df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
```

Insights : Above graphs shows most of the buyers are from IT sector,Healthcare and Aviation sector.

```python
df.head()
```

#### Marital_Status	Analysis -

```python
# orders count by Marital_Status-

sns.set(rc={'figure.figsize':(7,5)})
ax=sns.countplot(data=df,x='Marital_Status',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# sales by Marital_Status-

marital_sales=df.groupby('Marital_Status')['Amount'].sum()

plt.pie(marital_sales,labels=['Married','Unmarried'],autopct='%1.1f%%',startangle=90)

plt.title('Sales by Marital Status')
plt.show()
```

Insights: The above graphs shows that mostly buyers are single(female).

```python
df.head()
```

#### Zone Analysis-

```python
# orders count by Zone-

sns.set(rc={'figure.figsize':(7,5)})
ax=sns.countplot(data=df,x='Zone',color='magenta')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# sales by Zone-

sales_zone=df.groupby('Zone')['Amount'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=sales_zone,x='Zone',y='Amount',color='purple')
```

Insights: from the above graphs we can see that most of the orders and sales are from central,southern and western zone 

#### customer analysis by sales -

```python
# top 10 customers name by total sales-

top_customers = df.groupby('Cust_name')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()
sns.barplot(data=top_customers,x='Cust_name',y='Amount',color='brown')
```
